```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import wofz 
import scipy
from scipy.sparse import diags
```

```python
%matplotlib inline
import matplotlib
font = {'family' : 'sans-serif',
        'size'   : 14}

matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

alpha = np.linspace(0, 1.1, int(1e4))

ax.plot(alpha, 1-alpha, linewidth=4, color="tab:blue", label=r"$|V_{\perp, De}/V_{\perp, Dp}|$", alpha=0.7)
ax.plot(alpha, np.abs((1-alpha)/(1-4*alpha)), linewidth=4, color="tab:red", label=r"$|V_{\perp, DHe^{+}}/V_{\perp, Dp}|$", alpha=0.7)
ax.plot(alpha, np.abs((1-alpha)/(1-16*alpha)), linewidth=4, color="tab:green", label=r"$|V_{\perp, DO^{+}}/V_{\perp, Dp}|$", alpha=0.7)

ax.axvline(1, color="black", ls="--")
ax.axvline(1/4, color="black", ls="--")
ax.axvline(1/16, color="black", ls="--")

legend = ax.legend(ncols=1, fancybox=False, shadow=False,  loc="upper right")
legend.get_frame().set_alpha(0)

ax.set_xscale("log")

ax.set_xlabel(r"$\omega_{0}/\Omega_{cp}$")

ax.set_xticks([1/16, 1/4,  1])
ax.set_xticklabels([r"$1/16$", r"$1/4$", r"$1$"])

ax.axvspan(1/22, 1/16, color='tab:green', alpha=0.1)
ax.axvspan(1/16, 1/4, color='tab:red', alpha=0.1)
ax.axvspan(1/4, 1, color='tab:purple', alpha=0.1)
ax.axhline(1, color="black", ls="--")
ax.set_xlim(1/22, 1.1)

ax.set_ylim(0, 10)
ax.set_yticks([1, 3, 5, 7, 9])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.grid(alpha=0.2)
plt.tight_layout()

plt.savefig("figures/estimates/drift_estimate_emic.png", dpi=400)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

alpha = np.linspace(0, 1.1, int(1e4))
ax.plot(alpha, 1/(alpha - 1), linewidth=4, color="tab:purple", label=r"p", alpha=0.7)
ax.plot(alpha, 1/(alpha - 1/4), linewidth=4, color="tab:red", label=r"He$^{+}$", alpha=0.7)
ax.plot(alpha, 1/(alpha - 1/16), linewidth=4, color="tab:green", label=r"O$^{+}$", alpha=0.7)

ax.axvline(1, color="black", ls="--")
ax.axvline(1/4, color="black", ls="--")
ax.axvline(1/16, color="black", ls="--")


legend = ax.legend(ncols=1, fancybox=False, shadow=False,  loc="upper right")
legend.get_frame().set_alpha(0)


ax.set_xscale("log")

ax.set_xlabel(r"$\omega_{0}/\Omega_{cp}$")
ax.set_ylabel(r"$\propto U_{\perp, Di}$")

ax.set_xticks([1/16, 1/4,  1])
ax.set_xticklabels([r"$1/16$", r"$1/4$", r"$1$"])

ax.axvspan(1/25, 1/16, color='tab:green', alpha=0.1)
ax.axvspan(1/16, 1/4, color='tab:red', alpha=0.1)
ax.axvspan(1/4, 1, color='tab:purple', alpha=0.1)
ax.axhline(0, color="black", ls="--")
ax.set_xlim(1/25, 1.05)
ax.set_ylim(-5000, 5000)

ax.set_yticks([0])
ax.set_yscale("symlog", linthresh=1)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.grid(alpha=0.2)
plt.tight_layout()
plt.savefig("figures/estimates/U_drift_estimate.png", dpi=400)
```
