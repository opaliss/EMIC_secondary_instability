## Dispersion relation solver for MTSI


```python
# normalization
# time is normalized to the electron cyclotron frequency 
# space is normalized to electron inertial length d_e

# match the NASA proposal
# cold electron density
n_e = 1  # ne
# core proton density
n_p = 0.8  # ne
# beam proton density
n_O =  0.03 # ne

# mass ratios
mp_me = 1836 # dimensionless
mO_me = 1836 * 16 # dimensionless
omega_pe = 2 # wce
omega_ce = 1 # wce

# assume the plasma is isothermal Te=Tp
betae = 1e-4
alpha_e_par = np.sqrt(betae) # d_e x wce
alpha_p_par = alpha_e_par / np.sqrt(mp_me)  # d_e x wce
alpha_O_par = alpha_e_par / np.sqrt(mO_me)  # d_e x wce

rhop = alpha_e_par * np.sqrt(mp_me) / np.sqrt(2)

# relative drift
UD_p = -10 * (alpha_p_par/np.sqrt(2))
UD_O = 20 * (alpha_O_par/np.sqrt(2))
```

```python
print("current = ", UD_O*n_O + UD_p*n_p) 
print("(vbeam - vcore)/vti = ", (UD_O - UD_p)/(alpha_p_par/np.sqrt(2)))
```

```python
UD_O/ (alpha_p_par/np.sqrt(2))
```

```python
UD_p / (alpha_p_par/np.sqrt(2))
```

```python
def disp_k_(k_,
            theta_,
            n_p_=n_p,
            n_O_=n_O,
            omega_pe_=omega_pe,
            omega_ce_=omega_ce,
            mp_me_=mp_me,
            mO_me_=mO_me,
            UD_p_=UD_p,
            UD_O_=UD_O,
            alpha_e_par_=alpha_e_par,
            alpha_p_par_=alpha_p_par,
            alpha_O_par_=alpha_O_par,
            n_max_=10,
            electron_response_="hot"):
    if electron_response_=="hot":
        return lambda omega: 1 + electron_response(k_=k_, omega_=omega, omega_ce_=omega_ce_, omega_pe_=omega_pe_, theta_=theta_, alpha_c_par_=alpha_e_par_, n_max_=n_max_) \
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mp_me_), alpha_i_=alpha_p_par_, 
                                            k_=k_, theta_=theta_, v_0_=UD_p_, n_=n_p_, omega=omega)\
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mO_me_), alpha_i_=alpha_O_par_, 
                                    k_=k_, theta_=theta_, v_0_=UD_O_, n_=n_O_, omega=omega)
    elif electron_response_=="cold":
        return lambda omega: 1 + electron_response_cold(omega=omega, omega_ce_=omega_ce_, omega_pe_=omega_pe_, theta_=theta_) \
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mp_me_), alpha_i_=alpha_p_par_, 
                                            k_=k_, theta_=theta_, v_0_=UD_p_, n_=n_p_, omega=omega)\
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mO_me_), alpha_i_=alpha_O_par_, 
                                    k_=k_, theta_=theta_, v_0_=UD_O_, n_=n_O_, omega=omega)
```

# electron/Oxygen instability 

```python
k_vec = np.abs(np.linspace(0.1, 15, 40) /rhop) # d_e

theta_vec = (np.pi/2 - np.linspace(0, 4, 25) * np.pi / 180) # radians

omega_guess = 100 / mp_me  # wce

sol_hot = np.ones((len(k_vec), len(theta_vec)), dtype="complex128") * np.nan
sol_cold = np.ones((len(k_vec), len(theta_vec)), dtype="complex128") * np.nan
```

```python
for ii in range(len(k_vec)):
    for jj in range(len(theta_vec)):
        try:
            sol_cold[ii, jj] = solve_dispersion(func=disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="cold"),
                                                 x0=k_vec[ii] * UD_O * np.sin(theta_vec[jj]) * 0.9 + 1e-4j, 
                                                 x1=omega_guess + 1e-3j)
            error = abs(disp_k_(k_=k_vec[ii], theta_=theta_vec[jj],  electron_response_="cold")(sol_cold[ii, jj]))
            print("[cold] dispersion residual = ", error)
            if error > 1e-4:
                sol_cold[ii, jj] = np.nan
        except:
            print("[cold] An exception occurred!")
        try:
            sol_hot[ii, jj] = solve_dispersion(func=disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="hot"),
                                                x0=k_vec[ii] * UD_O * np.sin(theta_vec[jj]) * 0.9 + 1e-4j, 
                                                x1=omega_guess + 1e-3j)
            
            error = abs(disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="hot")(sol_hot[ii, jj]))
            print("[hot] dispersion residual = ", error)
            if error > 1e-4:
                sol_hot[ii, jj] = np.nan
        except:
            print("[hot] An exception occurred!")
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 3))
pos = ax[0].contourf(k_vec*rhop, theta_vec*180/np.pi, 
                  sol_cold.real.T * mp_me, vmin=30, vmax=180,  levels=np.linspace(30, 180, 25))

cbar = fig.colorbar(pos)
cbar.ax.set_ylabel(r'$\frac{\omega_{r}}{|\Omega_{cp}|}$', labelpad=20, rotation=0, fontsize=20)
ax[0].set_xlabel(r'$|\vec{k}|\rho_{p}$', fontsize=15)
ax[0].set_ylabel(r'$\theta$ [deg.]', fontsize=15)
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].set_title("cold electron response")


pos = ax[1].contourf(k_vec*rhop, theta_vec*180/np.pi, 
                  sol_hot.real.T * mp_me, vmin=30, vmax=180,  levels=np.linspace(30, 180, 25))
ax[1].set_title("hot electron response")
cbar = fig.colorbar(pos)
cbar.ax.set_ylabel(r'$\frac{\omega_{r}}{|\Omega_{cp}|}$', labelpad=20, rotation=0, fontsize=20)
ax[1].set_xlabel(r'$|\vec{k}|\rho_{p}$', fontsize=15)
ax[1].set_ylabel(r'$\theta$ [deg.]', fontsize=15)
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)

plt.tight_layout()
plt.savefig("figures/linear_theory/proton_oxygen/frequency_proton_" + str(n_O)+ ".png")
```
