## Dispersion relation solver for MTSI


```python
# normalization
# time is normalized to the electron cyclotron frequency 
# space is normalized to electron inertial length d_e

# match the NASA proposal
# cold electron density
ne = 1  # ne
# core proton density
np_core = 0.9  # ne
# beam proton density
np_beam =  0. # ne

# mass ratios
mp_me = 1836 # dimensionless
omega_pe = 2 # wce
omega_ce = 1 # wce

# assume the plasma is isothermal Te=Tp
betae = 1e-4
alpha_c_par = np.sqrt(betae) # d_e x wce
alpha_p_par = alpha_c_par / np.sqrt(mp_me)  # d_e x wce

rhoi = alpha_c_par * np.sqrt(mp_me) / np.sqrt(2)

# relative drift
V0 = 10 * (alpha_p_par/np.sqrt(2))
UD_core = - 7 * (alpha_p_par/np.sqrt(2)) # V0 / (-1 - np_core/np_beam)
UD_beam = 9 * (alpha_p_par/np.sqrt(2)) #V0 + UD_core
```

```python
print("current = ", UD_beam*np_beam + UD_core*np_core) 
print("(vbeam - vcore)/vti = ", (UD_beam - UD_core)/(alpha_p_par/np.sqrt(2)))
```

```python
UD_core/ (alpha_p_par/np.sqrt(2))
```

```python
UD_beam / (alpha_p_par/np.sqrt(2))
```

```python
def disp_k_(k_,
            theta_,
            np_core_=np_core,
            np_beam_=np_beam,
            omega_pe_=omega_pe,
            omega_ce_=omega_ce,
            mp_me_=mp_me,
            UD_core_=UD_core,
            UD_beam_=UD_beam,
            alpha_c_par_=alpha_c_par,
            alpha_p_par_=alpha_p_par,
            n_max_=10,
            electron_response_="hot"):
    if electron_response_=="hot":
        return lambda omega: 1 + electron_response(k_=k_, omega_=omega, omega_ce_=omega_ce_, omega_pe_=omega_pe_, theta_=theta_, alpha_c_par_=alpha_c_par_, n_max_=n_max_) \
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mp_me_), alpha_i_=alpha_p_par_, 
                                            k_=k_, theta_=theta_, v_0_=UD_core_, n_=np_core_, omega=omega)\
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mp_me_), alpha_i_=alpha_p_par_, 
                                    k_=k_, theta_=theta_, v_0_=UD_beam_, n_=np_beam_, omega=omega)
    elif electron_response_=="cold":
        return lambda omega: 1 + electron_response_cold(omega=omega, omega_ce_=omega_ce_, omega_pe_=omega_pe_, theta_=theta_) \
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mp_me_), alpha_i_=alpha_p_par_, 
                                            k_=k_, theta_=theta_, v_0_=UD_core_, n_=np_core_, omega=omega)\
                             - ion_response(omega_pi_=omega_pe_/np.sqrt(mp_me_), alpha_i_=alpha_p_par_, 
                                    k_=k_, theta_=theta_, v_0_=UD_beam_, n_=np_beam_, omega=omega)
```

# electron/ion beam instability 

```python
k_vec = np.abs(np.linspace(0.1, 15, 40) /rhoi) # d_e

theta_vec = np.sign(UD_beam) * (np.pi/2 - np.linspace(0, 4, 25) * np.pi / 180) # radians

omega_guess =40 / mp_me  # wce

sol_hot = np.ones((len(k_vec), len(theta_vec)), dtype="complex128") * np.nan
sol_cold = np.ones((len(k_vec), len(theta_vec)), dtype="complex128") * np.nan
```

```python
for ii in range(len(k_vec)):
    for jj in range(len(theta_vec)):
        try:
            sol_cold[ii, jj] = solve_dispersion(func=disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="cold"),
                                                 x0=k_vec[ii] * UD_beam * np.sin(theta_vec[jj]) * 0.9 + 1e-4j, 
                                                 x1=omega_guess + 1e-3j)
            error = abs(disp_k_(k_=k_vec[ii], theta_=theta_vec[jj],  electron_response_="cold")(sol_cold[ii, jj]))
            print("[cold] dispersion residual = ", error)
            if error > 1e-4:
                sol_cold[ii, jj] = np.nan
        except:
            print("[cold] An exception occurred!")
        try:
            sol_hot[ii, jj] = solve_dispersion(func=disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="hot"),
                                                x0=k_vec[ii] * UD_beam * np.sin(theta_vec[jj]) * 0.9 + 1e-4j, 
                                                x1=omega_guess + 1e-3j)
            
            error = abs(disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="hot")(sol_hot[ii, jj]))
            print("[hot] dispersion residual = ", error)
            if error > 1e-4:
                sol_hot[ii, jj] = np.nan
        except:
            print("[hot] An exception occurred!")
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 3))
pos = ax[0].contourf(k_vec*rhoi, theta_vec*180/np.pi, 
                  sol_cold.real.T * mp_me, vmin=30, vmax=180,  levels=np.linspace(30, 180, 25))

cbar = fig.colorbar(pos)
cbar.ax.set_ylabel(r'$\frac{\omega_{r}}{|\Omega_{ci}|}$', labelpad=20, rotation=0, fontsize=20)
ax[0].set_xlabel(r'$|\vec{k}|\rho_{i}$', fontsize=15)
ax[0].set_ylabel(r'$\theta$ [deg.]', fontsize=15)
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].set_title("cold electron response")


pos = ax[1].contourf(k_vec*rhoi, theta_vec*180/np.pi, 
                  sol_hot.real.T * mp_me, vmin=30, vmax=180,  levels=np.linspace(30, 180, 25))
ax[1].set_title("hot electron response")
cbar = fig.colorbar(pos)
cbar.ax.set_ylabel(r'$\frac{\omega_{r}}{|\Omega_{ci}|}$', labelpad=20, rotation=0, fontsize=20)
ax[1].set_xlabel(r'$|\vec{k}|\rho_{i}$', fontsize=15)
ax[1].set_ylabel(r'$\theta$ [deg.]', fontsize=15)
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)

plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 3))
pos = ax[0].contourf(k_vec*rhoi, theta_vec*180/np.pi, 
                  sol_cold.imag.T * mp_me, vmin=1E-4, vmax=14, levels=np.linspace(1e-4, 14, 25))

cbar = fig.colorbar(pos)
cbar.ax.set_ylabel(r'$\frac{\gamma}{|\Omega_{ci}|}$', labelpad=20, rotation=0, fontsize=20)
ax[0].set_xlabel(r'$|\vec{k}| \rho_{i}$', fontsize=15)
ax[0].set_ylabel(r'$\theta$ [deg.]', fontsize=15)
cbar.set_ticks([4, 8, 12])
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].set_title("cold electron response")


pos = ax[1].contourf(k_vec*rhoi, theta_vec*180/np.pi, 
                  sol_hot.imag.T * mp_me, vmin=1E-4, vmax=14, levels=np.linspace(1e-4, 14, 25))

cbar = fig.colorbar(pos)
cbar.ax.set_ylabel(r'$\frac{\gamma}{|\Omega_{ci}|}$', labelpad=20, rotation=0, fontsize=20)
ax[1].set_xlabel(r'$|\vec{k}| \rho_{i}$', fontsize=15)
ax[1].set_title("hot electron response")
ax[1].set_ylabel(r'$\theta$ [deg.]', fontsize=15)
cbar.set_ticks([4, 8, 12])
#cbar.set_ticklabels(["$10^{-4}$", "0.1", "0.2"])
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)


plt.tight_layout()
plt.savefig("figures/linear_theory/gary/growth_rate_gary_electron_beam_" + str(np_core)+ ".png")
```

# Core instability 

```python
k_vec = np.abs(UD_beam/UD_core * k_vec) # d_e

omega_guess = 50 / mp_me # Omega_ce

sol_hot = np.ones((len(k_vec), len(theta_vec)), dtype="complex128") * np.nan
sol_cold = np.ones((len(k_vec), len(theta_vec)), dtype="complex128") * np.nan
```

```python
for ii in range(len(k_vec)):
    for jj in range(len(theta_vec)):
        try:
            sol_cold[ii, jj] = solve_dispersion(func=disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="cold"),
                                                x0=k_vec[ii] * UD_core * np.sin(theta_vec[jj]) * 0.9 + 1e-2j,
                                                x1=-omega_guess + 1e-2j)
            
            error = abs(disp_k_(k_=k_vec[ii], theta_=theta_vec[jj],  electron_response_="cold")(sol_cold[ii, jj]))
            print("[cold] dispersion residual = ", error)
            if error > 1e-3:
                sol_cold[ii, jj] = np.nan
        except:
            print("[cold] An exception occurred!")
        try:
            sol_hot[ii, jj] = solve_dispersion(func=disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="hot", n_max_=3),
                                                x0=k_vec[ii] * UD_core * np.sin(theta_vec[jj]) * 0.9 + 1e-2j,
                                                x1=-omega_guess + 1e-2j)
            
            error = abs(disp_k_(k_=k_vec[ii], theta_=theta_vec[jj], electron_response_="hot")(sol_hot[ii, jj]))
            print("[hot] dispersion residual = ", error)
            if error > 1e-4:
                sol_hot[ii, jj] = np.nan
        except:
            print("[hot] An exception occurred!")
```
